# Qualia Investments

Data overview:<br>
- Data was collected on 2019-09-17<br>
- Dataset includes 124 upcoming events at Madison Square Garden<br>
- See field tm_offers for current availability on Ticketmaster. Each element of tm_offers is a set of seats that are available at a specific price point. See tm_offers.offers.places for the seats and section.<br>
- Field sh_listings.listing contains the current stubhub listings.<br>

```python
# Importing Packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset

data=pd.read_json("madison_square_garden_tm_sh_data-2019-09-17.json")
```

# Exploratory Analysis

## Extracting data from TM for first event

```python
l=data['tm_offers'][0]
```

```python
import itertools
# First Event, where is a particular event
df = pd.DataFrame(columns=['event','faceValue','listPrice','totalPrice','noChargesPrice','section','row','seat','row_count'])
keys=['faceValue','listPrice','totalPrice','noChargesPrice']
keys2=['section','row','seat','row_count']

c=0
t=0
for l in data.iterrows():
    venue=[l[1]['EventName']]
    l=data['tm_offers'][t]
    for i in l:
        pricing=[i['offer'][key] for key in keys]
        for p in i['places']:
            seating=[p[key] for key in keys2]
            s=itertools.chain(venue,pricing,seating)
            df.loc[c]=list(s)
            c+=1
    print(venue)
    t+=1
    if t==2:
        break
```
